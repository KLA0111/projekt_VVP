Obecny postup pro hledani cesty v danem bludiste:

```python
maze_test = load_maze("data/maze_3.csv")
show_maze(maze_test)
path = get_path(maze_test)
show_maze(maze_test, path)
```

Postup fungovani get_path vypada takto: nejprve udelame matice incidence, pak projdeme ji celou, pak najdeme nejkrasti cestu z verhniho leveho vrcholu fo dolniho praveho.

```python
incidence = get_incindence(maze_test)
d = dijkstra(incidence)
path = get_path_dijkstra(d, incidence)
show_maze(maze_test, path)
```

Muzeme ukazat matice incindecnosti pro dane bludiste:

```python
incidence = get_incindence(maze_test)

fig, ax = plt.subplots()
ax.imshow(incidence, cmap='binary')
plt.show()
```
